```python
from IPython.core.display import HTML
def table_styling():
    return HTML("""
    <style>
    table.python_table {
         margin: 0          !important;
        padding: 0          !important;
    }
    table.python_table td, table.python_table th  {
        border-spacing: 0   !important;
        vertical-align: top !important;
        padding: 5px;
        border: 1px solid #ccc;
        font-family: "Lucida Grande";
    }
    table.python_table th {
        text-align: center;
    }
    
    table.python_table td > table td, table.python_table td > table tr, table.python_table td > table th {
        border: 1px solid #ddd;
    }
    
    </style>
    
    """)
def to_table(datum, **kwargs):
    title = ''
    colspan = str(len(datum)) if kwargs.get('axis') == 1 else '1'
    if kwargs.get('title'):
        title = '<tr><th colspan='+colspan+'>'+kwargs.get('title')+'</th></tr>'
    if kwargs.get('axis') == 1:
        cols = ["<td>"+ i +"</td>" for i in datum]
        return "<table class='python_table' cellpadding='0'><thead>"+title+"</thead><tbody><tr>" + "".join(cols) + "</tr></tbody></table>"
    else:
        rows = ["<tr><td>"+ i +"</td></tr>" for i in datum]
        return "<table class='python_table' cellpadding='0'><thead>"+title+"</thead><tbody>" + "".join(rows) + "</tbody></table>"
def show_in_table(datum, **kwargs):
    return HTML(to_table(datum, **kwargs))
table_styling()
```

```python
batch_size = 100
learning_rate = 0.001 / batch_size
encoder = rsm.RSM(momentum = 0.001, data = np.zeros([0,len(lexicon.items())]), hidden_units = 200, learning_rate=learning_rate)
errors = np.zeros(0)
```

Create the stochastic batch element with 100 elements per mini-batch:

```python
rc = utils.ResourceConverter(lexicon = lexicon)
batch = Batch(
    data=utils.mongo_database_global['restaurants'].find(), # from Mongo's cursor enumerator
    batch_size = batch_size,  # mini-batch
    shuffle = True, # stochastic
    conversion = rc.process # convert to matrices using lexicon)
)
```

Start mini-batch learning for 1000 epochs:

```python
epochs = 200
batch.batch_size = 100
new_errors = np.zeros(epochs)
start_time = time.time()
encoder.learning_rate = 0.001

for epoch in range(epochs):
    if epoch > 0 and epoch % 200 == 0:
        encoder.k = int(1 + (float(epoch) / float(epochs)) * 4)
    if epoch > 0 and epoch % 300 == 0:
        encoder.learning_rate = max(0.00001, encoder.learning_rate.get_value()*0.95)
    encoder.data = batch.next()
    new_errors[epoch] = encoder.train()
    if epoch > 0 and epoch % 500 == 0:
        encoder.save("backup.pkz")
    if epoch > 0 and epoch % 10 == 0:
        print("Epoch[%2d] : PPL = %.02f [# Gibbs steps=%d] elapsed = %.05fmn" % (epoch, new_errors[epoch],encoder.k, (time.time() - start_time)/60.0))
errors = np.append(errors, new_errors)
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(np.log(errors[3:170]))
axes.set_ylabel("Log perplexity")
axes.set_xlabel("Epochs")
axes.set_title("Training error for Stochastic Gradient Descent of RSM")
fig.savefig("plt.svg")
```
